```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

1.importing csv file

```python
df = pd.read_csv("Amazon Sale Report.csv", encoding="ISO-8859-1")
df.head(10)
```

2.Checking Null Values

```python
df.isnull().sum()
print(df)
```

3.Removing Rows will null values

```python
df_cleaned = df.dropna()
print("\nDataset after Removing Null Values:")
print(df_cleaned.isnull().sum())
```

4. Checking for duplicate rows

```python
duplicates = df_cleaned.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")
df_cleaned = df_cleaned.drop_duplicates()
print(f"\nDataset Shape after Removing Duplicates: {df_cleaned.shape}")
```

```python
# Final Cleaned Dataset
print("\nFinal Cleaned Dataset:")
print(df_cleaned.info())
```

```python
if df_cleaned.empty:
    print("\nThe cleaned DataFrame is empty! Please review the cleaning steps.")
else:
    # Step 7: Save the cleaned dataset to a new CSV file
    cleaned_file_path = "Cleaned_Amazon_Sale_Report.csv"
    df_cleaned.to_csv(cleaned_file_path, index=False)
    print(f"\nCleaned dataset saved to: {cleaned_file_path}")

# Step 8: Final confirmation
print("\nFinal Cleaned Data Overview:")
print(df_cleaned.info())
print("\nSample Data:")
print(df_cleaned.head())
```
