```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Amazon Sale Report.csv", encoding="ISO-8859-1")
df.head(10)
```

```python
df.isnull().sum()
print(df) 
df_cleaned = df.dropna()
print("\nDataset after Removing Null Values:")
print(df_cleaned.isnull().sum()) 
duplicates = df_cleaned.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")
df_cleaned = df_cleaned.drop_duplicates()
print(f"\nDataset Shape after Removing Duplicates: {df_cleaned.shape}") 
```

```python
# Final Cleaned Dataset
print("\nFinal Cleaned Dataset:")
print(df_cleaned.info())
```

```python
df.drop('New', inplace=True, axis=1) 
df.drop('PendingS', inplace=True, axis=1) 
```

```python
df.head(-100)
```

```python
print("Dataset Columns:", df.columns)
```

```python
# Order Status Analysis 
status_counts = df['Status'].value_counts(normalize=True) * 100
print("\nProportion of Order Statuses (%):")
print(status_counts)
# Identify problematic statuses
problematic_statuses = ['Cancelled', 'Returned', 'Delayed']  # Define status keywords for problematic cases
df['Problematic'] = df['Status'].apply(lambda x: any(status in x for status in problematic_statuses))

# Calculate proportion of problematic orders
problematic_proportion = df['Problematic'].mean() * 100
print(f"\nProportion of Problematic Orders: {problematic_proportion:.2f}%")

#Courier Status Patterns
courier_status_counts = df['Courier Status'].value_counts()
print("\nCourier Status Patterns:")
print(courier_status_counts)
```

```python
 # Bar Graph Proportion of Order Statuses
plt.figure(figsize=(10, 8))
status_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Proportion of Order Statuses", fontsize=12)
plt.xlabel("Order Status", fontsize=10)
plt.ylabel("Percentage", fontsize=10)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart Problematic vs. Non-Problematic Orders
problematic_summary = df['Problematic'].value_counts(normalize=True) * 100
plt.figure(figsize=(12, 8))
plt.pie(
    problematic_summary, 
    labels=['Non-Problematic Orders', 'Problematic Orders'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['green', 'red']
)
plt.title("Proportion of Problematic Orders", fontsize=16)
plt.legend()
plt.show()
```

```python
# Bar Graph Courier Status Distribution
plt.figure(figsize=(8, 4))
courier_status_counts.plot(kind='bar', color='blue', edgecolor='black')
plt.title("Courier Status Distribution", fontsize=16)
plt.xlabel("Courier Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#  Histogram Distribution of Amounts for Problematic Orders
problematic_orders = df[df['Problematic']]
plt.figure(figsize=(10, 6))
plt.hist(problematic_orders['Amount'], bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Amounts for Problematic Orders", fontsize=16)
plt.xlabel("Amount", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
