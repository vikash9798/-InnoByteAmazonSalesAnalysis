```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Amazon Sale Report.csv", encoding="ISO-8859-1")
df.head(10)
```

```python
df.isnull().sum()
print(df)
```

```python
df_cleaned = df.dropna()
print("\nDataset after Removing Null Values:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")
df = df.drop_duplicates()
print(f"\nDataset Shape after Removing Duplicates: {df.shape}")
```

```python
print("Columns in Dataset:", df.columns)
```

```python
df.drop('New', inplace=True, axis=1) 
df.drop('PendingS', inplace=True, axis=1) 
df.head()
```

```python

# Validate required columns
required_columns = ['Fulfilment', 'Courier Status', 'Date']
for col in required_columns:
    if col not in df.columns:
        raise KeyError(f"Column '{col}' not found in the dataset!")

# Map 'Courier Status' to numeric values to estimate 'Delivery Time'
status_mapping = {
    'Shipped': 2,
    'On the Way': 5,
    'Delivered': 1,
    'Cancelled': np.nan  # Exclude cancelled orders
}
df['Delivery Time'] = df['Courier Status'].map(status_mapping)
print(f"\nNon-NaN Delivery Time values: {df['Delivery Time'].count()}")

# Assume Customer Satisfaction is inversely related to Delivery Time
df['Customer Satisfaction'] = np.where(df['Delivery Time'] <= 3, 1, 0)

# Drop rows where critical fields are missing
df = df.dropna(subset=['Fulfilment', 'Delivery Time', 'Customer Satisfaction'])
print(f"\nDataset after cleaning: {df.shape[0]} rows")

if df.empty:
    raise ValueError("No valid data available after cleaning!")
```

```python
#Analyze Average Delivery Time by Fulfillment Method
fulfillment_delivery = df.groupby('Fulfilment')['Delivery Time'].mean().sort_values()

print("\nAverage Delivery Time by Fulfillment Method:")
print(fulfillment_delivery)

# Analyze Average Customer Satisfaction by Fulfillment Method
fulfillment_satisfaction = df.groupby('Fulfilment')['Customer Satisfaction'].mean().sort_values()

print("\nAverage Customer Satisfaction by Fulfillment Method:")
print(fulfillment_satisfaction)
```

```python
 # Visualization 1: Average Delivery Time by Fulfillment Method
plt.figure(figsize=(10, 6))
fulfillment_delivery.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Average Delivery Time by Fulfillment Method", fontsize=16)
plt.xlabel("Fulfillment Method", fontsize=12)
plt.ylabel("Average Delivery Time (Days)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Customer Satisfaction by Fulfillment Method
plt.figure(figsize=(12, 10))
plt.pie(
    fulfillment_satisfaction, 
    labels=fulfillment_satisfaction.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title("Customer Satisfaction by Fulfillment Method", fontsize=16)
plt.legend(loc= 'upper left', fontsize='8')
plt.show()
```

```python

# Delivery Time Trends
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])

monthly_delivery = df.groupby([df['Date'].dt.to_period('M'), 'Fulfilment'])['Delivery Time'].mean().unstack()

plt.figure(figsize=(14, 12))
monthly_delivery.plot(kind='line', marker='o')
plt.title("Monthly Delivery Time Trends by Fulfillment Method", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Delivery Time (Days)", fontsize=12)
plt.grid(True)
plt.legend(title='Fulfillment Method')
plt.tight_layout()
plt.show()
```

Why This Analysis?
Optimize Fulfillment Strategy:

Identifies the most efficient fulfillment methods.
Helps businesses prioritize methods that minimize delivery times.
Enhance Customer Experience:

Provides insights to improve service quality and customer retention.
