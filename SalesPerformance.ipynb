importing all library

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

1.Sales performance Analysis

```python
df = pd.read_csv("Amazon Sale Report.csv", encoding="ISO-8859-1")
df.head(10)
```

checking Null values

```python
df.isnull().sum()
print(df)
```

```python
df_cleaned = df.dropna()
print("\nDataset after Removing Null Values:")
print(df_cleaned.isnull().sum())
```

```python
duplicates = df_cleaned.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")
df_cleaned = df_cleaned.drop_duplicates()
print(f"\nDataset Shape after Removing Duplicates: {df_cleaned.shape}")
```

```python
print("\nFinal Cleaned Dataset:")
print(df_cleaned.info())
```

```python
# Ensure the 'Amount' column is numeric 
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
```

```python
# Summary Statistics
print("\nSummary Statistics of Sales Performance:")
sales_stats = {
    "Total Sales Amount Rs": df['Amount'].sum(),
    "Average Sales Amount Rs": df['Amount'].mean(),
    "Median Sales Amount Rs": df['Amount'].median(),
    "Standard Deviation Amount Rs": df['Amount'].std(),
    "Maximum Sales Amount Rs": df['Amount'].max(),
    "Minimum Sales Amount Rs": df['Amount'].min(),
}
for key, value in sales_stats.items():
    print(f"{key}: {value:.2f}")
```

```python
# Sales Over Time 
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])  # Drop invalid dates 
print(df)
# Aggregate sales by month
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Amount'].sum()
print(monthly_sales)
```

```python
# Visualize Sales Trends as a Bar Graph
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar', color='blue', edgecolor='black')
plt.title("Sales Performance Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(["Total Sales"], loc='upper left', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Sales Distribution
plt.figure(figsize=(10, 5))
plt.hist(df['Amount'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Sales", fontsize=16)
plt.xlabel("Sales", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Top 10 Sales Days
top_sales_days = df.nlargest(10, 'Amount')[['Date', 'Amount']]
print("\nTop 10 Sales Days:")
print(top_sales_days)
plt.figure(figsize=(14, 12))
plt.pie(
    top_sales_days['Amount'], 
    labels=top_sales_days['Date'].dt.strftime('%Y-%m-%d'),  # Formatting date labels
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title("Top 10 Sales Days - Distribution", fontsize=16)
plt.legend()
plt.show()
```

```python
# Using NumPy for Additional Analysis
sales_array = df['Amount'].dropna().values  # Ensure no NaN values in the data

percentiles = {
    "25th Percentile": np.percentile(sales_array, 25),
    "Median (50th Percentile)": np.percentile(sales_array, 50),
    "75th Percentile": np.percentile(sales_array, 75),
    "Variance": np.var(sales_array),
    "Coefficient of Variation": np.std(sales_array) / np.mean(sales_array),
}
print("\nNumPy Analysis:")
for key, value in percentiles.items():
    print(f"{key}: {value:.2f}")
keys = list(percentiles.keys())
values = list(percentiles.values())
if len(values) > 0 and all(np.isfinite(values)):
    plt.figure(figsize=(10, 6))
    bars = plt.bar(keys, values, color=['blue', 'green', 'orange', 'red', 'purple'])

    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f'{height:.2f}',
            ha='center',
            va='bottom',
            fontsize=10
        )

    plt.title("NumPy Analysis of Sales Data", fontsize=16)
    plt.ylabel("Value", fontsize=12)
    plt.xlabel("Metrics", fontsize=12)
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for clarity
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(["Key Metrics"], loc='upper right', fontsize=10)

    plt.tight_layout()
    plt.show()
else:
    print("Invalid or insufficient data to plot the bar graph.")
```

```python
#  Line Graph of  Sales Performance Over Time
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='blue', label='Monthly Sales')
plt.title("Sales Performance Over Time", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
# Histogram Distribution of Sales
plt.figure(figsize=(10, 5))
plt.hist(df['Amount'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Sales", fontsize=16)
plt.xlabel("Sales", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()

# Pie Chart Top 10 Sales Days
plt.figure(figsize=(14, 12))
plt.pie(
    top_sales_days['Amount'], 
    labels=top_sales_days['Date'].dt.strftime('%Y-%m-%d'),
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title("Top 10 Sales Days - Distribution", fontsize=16)
plt.legend(loc='upper right', fontsize='8')
plt.show()

# Bar Graph with NumPy Analysis
keys = list(percentiles.keys())
values = list(percentiles.values())
plt.figure(figsize=(10, 6))
bars = plt.bar(keys, values, color=['blue', 'green', 'orange', 'red', 'purple'])
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.2f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

# Adding title and labels
plt.title("NumPy Analysis of Sales Data", fontsize=16)
plt.ylabel("Value", fontsize=12)
plt.xlabel("Metrics", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(["Key Metrics"], loc='upper right', fontsize=10)
plt.tight_layout()
plt.show()
```
